```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
```

# Input data

```python
# height (cm)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
# weight (kg)
y = np.array([ 49, 50, 51, 54, 58, 59, 60, 62, 63, 64, 66, 67, 68])
```

# Biểu diễn dữ liệu

```python
plt.xlabel("Height(cm)")
plt.ylabel("Weight(kg)")
plt.scatter(X, y, color='r')
plt.axis([140, 190, 45, 75])
plt.show()
```

# Tính tham số của mô hình

```python
Xbar = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
w = np.dot(np.linalg.pinv(np.dot(Xbar.T, Xbar)), np.dot(Xbar.T, y))
print(w)
```

# Dự đoán người có chiều cao 173.8cm

```python
print("Chiều cao dự đoán là:", w[0] + 173.8*w[1])
```

```python
plt.xlabel("Height(cm)")
plt.ylabel("Weight(kg)")
plt.scatter(X, y, color='r')
plt.axis([140, 190, 45, 75])
x0 = np.linspace(145, 185, 2)
y0 = w[0] + w[1]*x0
plt.plot(x0,y0)
plt.show()
```
